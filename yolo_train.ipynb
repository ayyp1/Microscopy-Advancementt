```python
from ultralytics import YOLO
import matplotlib.pyplot as plt 
import numpy as np 

model = YOLO("yolo11n.pt")
```

```python
result = model.predict("image.HEIC")
```

```python
%matplotlib inline
plt.imshow(np.squeeze(result[0].plot()))
plt.show()
```

```python
results = model.train(
    data="yolo.yaml",  # Path to your YAML file
    epochs=30,          # Number of epochs
    imgsz=640,          # Image size (adjust based on your data)
    batch=8,           # Batch size (adjust based on your GPU memory)
    name="n_model"   # Name of the training run
)
```

```python
import matplotlib.pyplot as plt

# Example: Precision-Recall Curve
pr_curve = results.curves_results[0]
x = pr_curve[0]  # Recall
y_giardia = pr_curve[1][0]  # Precision for Giardia
y_crypto = pr_curve[1][1]  # Precision for Cryptosporidium

plt.plot(x, y_giardia, label="Giardia")
plt.plot(x, y_crypto, label="Cryptosporidium")
plt.xlabel("Recall")
plt.ylabel("Precision")
plt.title("Precision-Recall Curve")
plt.legend()
plt.show()
```
